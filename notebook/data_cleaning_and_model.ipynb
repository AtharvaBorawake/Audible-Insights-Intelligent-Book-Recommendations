```python
import re
import numpy as np

def convert_to_minutes(time_str):
    
    if pd.isna(time_str):
        return np.nan
    
    hours = 0
    minutes = 0
    
    h = re.search(r'(\d+)\s*hour', time_str)
    m = re.search(r'(\d+)\s*minute', time_str)
    
    if h:
        hours = int(h.group(1))
        
    if m:
        minutes = int(m.group(1))
        
    return hours*60 + minutes
```

```python
df["Listening Time (minutes)"] = df["Listening Time"].apply(convert_to_minutes)
```

```python
def extract_genre(text):
    
    if pd.isna(text):
        return np.nan
    
    genres = re.findall(r'in ([^,(]+)', text)
    
    genres = [g.strip() for g in genres]
    
    # remove platform category
    genres = [g for g in genres if g != "Audible Audiobooks & Originals"]
    genres = [g for g in genres if g != "Audible Audiobooks & Originals)"]
    
    return ", ".join(genres) if genres else np.nan
```

```python
def recommend_cluster_books(title, n=5):

    idx = indices[title]

    cluster_id = df.iloc[idx]["cluster"]

    cluster_books = df[df["cluster"] == cluster_id]

    cluster_books = cluster_books[cluster_books.index != idx]

    return cluster_books[["Book Name","Author","Rating","Genre"]].head(n)
```

```python
def recommend_hybrid_books(title, n=5):

    idx = indices[title]

    cluster_id = df.iloc[idx]["cluster"]

    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # skip the first book (itself)
    book_indices = [i[0] for i in sim_scores[1:]]

    filtered_books = df.iloc[book_indices]

    filtered_books = filtered_books[filtered_books["cluster"] == cluster_id]

    return filtered_books[["Book Name","Author","Rating","Genre"]].head(n)
```

```python
# Calculate the mean rating across the whole dataset
C = df['Rating'].mean()

# Calculate the minimum reviews required to be 'reliable' (e.g., 75th percentile)
m = df['Number of Reviews'].quantile(0.75)

def weighted_rating(x, m=m, C=C):
    v = x['Number of Reviews']
    R = x['Rating']
    # IMDb formula
    return (v/(v+m) * R) + (m/(m+v) * C)

# Apply the formula to create a new score column
df['weighted_score'] = df.apply(weighted_rating, axis=1)
```

```python
def recommend_quality_hybrid(title, n=5):
    # 1. Get the book index
    idx = indices[title]
    
    # 2. Get similarity scores for all books
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # 3. Filter books in the same cluster (Optional but improves precision)
    cluster_id = df.iloc[idx]["cluster"]
    
    # Create a list of (index, similarity_score)
    # We only keep books from the same cluster to ensure high precision
    filtered_scores = []
    for i, score in sim_scores:
        if df.iloc[i]["cluster"] == cluster_id and i != idx:
            # We combine the similarity score with the quality score
            # You can tune the weights (e.g., 0.5 each)
            total_score = (score * 0.5) + (df.iloc[i]['weighted_score'] / 5.0 * 0.5)
            filtered_scores.append((i, total_score))
            
    # 4. Sort by the combined score
    filtered_scores = sorted(filtered_scores, key=lambda x: x[1], reverse=True)
    
    # 5. Return top N
    book_indices = [i[0] for i in filtered_scores[:n]]
    return df[["Book Name", "Author", "Rating", "Genre", "weighted_score"]].iloc[book_indices]
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error

# --- 1. PRE-CHECK: Ensure we don't have NaNs in the weighted score calculation ---
# (Run this if you haven't already to make sure 'weighted_score' is valid)
C = df['Rating'].mean()
m = df['Number of Reviews'].quantile(0.75)

def weighted_rating(x, m=m, C=C):
    v = x['Number of Reviews']
    R = x['Rating']
    # If rating is NaN, return the global mean (C)
    if pd.isna(R): return C
    return (v/(v+m) * R) + (m/(v+m) * C)

df['weighted_score'] = df.apply(weighted_rating, axis=1)

# --- 2. UPDATED EVALUATION FUNCTION (NaN-Safe) ---
def run_final_evaluation_safe(model_list, k=5, sample_size=50):
    results = []
    
    # Filter sample to only include books that have a valid Rating
    valid_books = df[df['Rating'].notna() & df['Book Name'].isin(indices.keys())]
    sample_titles = valid_books.sample(min(sample_size, len(valid_books)))["Book Name"].values

    for name, func in model_list:
        t_prec, t_rec, t_rmse, count = 0, 0, 0, 0
        
        for title in sample_titles:
            try:
                # Use n=k to avoid positional argument errors
                recs = func(title, n=k) 
                
                if recs is None or recs.empty:
                    continue
                
                s_idx = indices[title]
                s_rating = df.iloc[s_idx]["Rating"]
                
                relevant_count = 0
                rec_ratings = []
                
                for r_title in recs["Book Name"]:
                    r_idx = indices[r_title]
                    r_rating = df.iloc[r_idx]["Rating"]
                    
                    # ONLY use the recommendation if the rating is NOT NaN
                    if pd.notna(r_rating):
                        rec_ratings.append(r_rating)
                        if is_relevant(s_idx, r_idx):
                            relevant_count += 1
                
                # Only update totals if we actually got valid rated recommendations
                if len(rec_ratings) > 0:
                    t_prec += (relevant_count / k)
                    t_rec += (relevant_count / 10)
                    
                    # Calculate RMSE only on the valid ratings found
                    mse = mean_squared_error([s_rating] * len(rec_ratings), rec_ratings)
                    t_rmse += np.sqrt(mse)
                    count += 1
                    
            except Exception as e:
                # Skip errors for specific books
                continue
            
        results.append({
            "Model": name,
            "Precision@K": t_prec/count if count > 0 else 0,
            "Recall@K": t_rec/count if count > 0 else 0,
            "RMSE": t_rmse/count if count > 0 else 0
        })
        
    return pd.DataFrame(results)

# --- 3. EXECUTE COMPARISON ---
my_models = [
    ("Content-Based", recommend_books),
    ("Cluster-Based", recommend_cluster_books),
    ("Hybrid Model", recommend_hybrid_books),
    ("Quality-Hybrid (NEW)", recommend_quality_hybrid)
]

evaluation_summary = run_final_evaluation_safe(my_models)
print(evaluation_summary)
```
